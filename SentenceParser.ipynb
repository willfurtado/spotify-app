```python
sentence = """Tado, or Will Furtado, is an independent musician and lyricist hailing from Berkeley, California. 
Growing up in a small coastal region of Massachusetts, Tado cites his brief stint with the Dem Roots rap crew to 
have catalyzed his passion for music. Although the majority of his discography remains unreleased, the most 
passionate fans are known to have access to recordings from his early career. Tado's highly-anticipated debut 
release, Scooby, was a collaborative project with his fellow roommate and friend, Jinho Shin. Superficially, the 
track details unrequited love; on a more profound level, it raises the moral dilemma of choosing between what one 
is truly passionate about, and what one knows will give them economic success.""".replace("\n", "")
```

# NLTK

```python
import nltk
from nltk.collocations import *
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(sentence.split())
finder.apply_freq_filter(1)

for i in finder.score_ngrams(bigram_measures.likelihood_ratio):
    print (i)
```

```python
trigram_measures = nltk.collocations.TrigramAssocMeasures()
triFinder = TrigramCollocationFinder.from_words(sentence.split())
triFinder.apply_freq_filter(1)

for i in triFinder.score_ngrams(trigram_measures.likelihood_ratio):
    print (i)
```

# Spacy

```python
import spacy
from spacy.lang.en import English
nlp = spacy.load("en_core_web_sm")
```

```python
def get_chunk(doc, pos):
    "Function to get PPs from a parsed document."
    pps = []
    for token in doc:
        # Try this with other parts of speech for different subtrees.
        if token.pos_ == pos:
            pp = ' '.join([tok.orth_ for tok in token.subtree])
            pps.append(pp)
    return pps
```

```python
doc = nlp(sentence)
pps = get_chunk(doc, "ADP")
entities=[str(i) for i in doc.ents]
noun_chunks = [chunk.text for chunk in doc.noun_chunks]
print(f"SUBJECTS:\n {noun_chunks}\n")
print(f"ENTITIES:\n {entities}\n")
print(f"PREPOSITIONAL PHRASES:\n {pps}\n")
```
